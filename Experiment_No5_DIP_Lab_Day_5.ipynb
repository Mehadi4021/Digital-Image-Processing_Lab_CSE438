# Problem: 1
b)	Write a program to read a gray scale image and calculate itâ€™s mean intensity value. Now, increase the intensity value by a constant range and calculate the new mean intensity value.

```python

# Read the grayscale image (Replace 'your_image.jpg' with the actual image file name)
image = cv2.imread('/content/drive/MyDrive/DIP_Lab/Day_4/pizza.png', cv2.IMREAD_GRAYSCALE)

# Calculate the original mean intensity
original_mean = np.mean(image)

# Increase intensity by a constant (e.g., 50)
increase_value = 50
brightened_image = cv2.add(image, increase_value)

# Calculate the new mean intensity
new_mean = np.mean(brightened_image)


```

```python
# Display results
print(f"Original Mean Intensity: {original_mean}")
print(f"New Mean Intensity after increasing by {increase_value}: {new_mean}")

# Plot images side by side
plt.figure(figsize=(10, 5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis("off")

# Brightened Image
plt.subplot(1, 2, 2)
plt.imshow(brightened_image, cmap='gray')
plt.title(f"Brightened Image (+{increase_value})")
plt.axis("off")

plt.show()
```

# Negation of Image

```python
# Provide the path to your image in Google Drive
image_path = "/content/drive/MyDrive/DIP_Lab/Day_4/clock.png"  # Change this to your image path

# Read the image in grayscale
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Perform image negation
negated_image = cv2.bitwise_not(image)

# Display original and negated images side by side
plt.figure(figsize=(10, 5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis("off")

# Negated Image
plt.subplot(1, 2, 2)
plt.imshow(negated_image, cmap='gray')
plt.title("Negated Image")
plt.axis("off")

plt.show()
```

# Shearing in X axis

Shearing is a transformation that skews the image along an axis.
Here, we are applying shearing along the x-axis, meaning that each row of pixels will be shifted horizontally based on its y-coordinate.

An affine transformation is a type of geometric transformation that preserves points, straight lines, and planes. In simple terms, it is a transformation that involves operations such as translation, scaling, rotation, shearing, and reflection, but it does not distort shapes like bending or curving. After applying an affine transformation, the resulting image will still have parallel lines parallel and ratios of distances between points preserved.

Shearing is a geometric transformation that slants the shape of an object. In the case of shearing along the x-axis, points in the image are displaced horizontally in proportion to their y-coordinates while keeping the y-coordinates unchanged.

An affine transformation is a linear transformation that preserves parallelism between lines but not necessarily angles or lengths. It includes transformations like:

âœ… Translation (Shifting)
âœ… Scaling (Resizing)
âœ… Rotation (Turning)
âœ… Shearing (Slanting)

```python
# Get the image dimensions
rows, cols = image.shape[:2]
```

```python
# Define the shear matrix for shearing along the x-axis
shear_factor = 0.5  # Adjust this factor as needed (0.1 to 1)
M = np.float32([[1, shear_factor, 0], [0, 1, 0]]) #This is a 2Ã—3 affine transformation matrix used for shearing.


# Apply the affine transformation (shearing)
sheared_img_x = cv2.warpAffine(image, M, (cols, rows))
```

```python
# Display original and negated images side by side
plt.figure(figsize=(10, 5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis("off")

# Sheared in x axis Image
plt.subplot(1, 2, 2)
plt.imshow(sheared_img_x, cmap='gray')
plt.title("Sheared Image in x axis")
plt.axis("off")

plt.show()
```

# Shearing in y axis

```python
# Define the shear matrix for Y-axis shearing
shear_factor = 0.5  # Adjust this value as needed
M = np.float32([[1, 0, 0], [shear_factor, 1, 0]])

# Apply the affine transformation (shearing in Y-axis)
sheared_img_y = cv2.warpAffine(image, M, (cols, rows))



```

```python
# Display original and negated images side by side
plt.figure(figsize=(10, 5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis("off")

# Sheared in x axis Image
plt.subplot(1, 2, 2)
plt.imshow(sheared_img_y, cmap='gray')
plt.title("Sheared Image in y axis")
plt.axis("off")

plt.show()
```
