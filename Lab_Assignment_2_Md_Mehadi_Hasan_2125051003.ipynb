Import necessary libraries

```python
from google.colab import files
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import urllib.request
import numpy as np
```

#### Read the image

```python
# Step 1: Provide the URL of the image
image_url = 'https://media.istockphoto.com/id/1353379172/photo/cute-little-african-american-girl-looking-at-camera.jpg?s=612x612&w=0&k=20&c=RCOYytwS2nMGfEb80oyeiCcIiqMQu6wnTluAaxMBye4='  # Replace with the actual URL
```

```python
# Step 2: Download the image using urllib
with urllib.request.urlopen(image_url) as response:
    img_array = np.asarray(bytearray(response.read()), dtype=np.uint8)  # Convert to NumPy array
```

```python
# Step 3: Decode the image (convert it into OpenCV format)
img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)  # Decode as a color image (BGR format)
```

```python
# Step 4: Convert BGR to RGB for correct display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python

# Step 5: Display the image using Matplotlib
plt.imshow(img_rgb)
plt.axis('off')  # Remove axes for cleaner display
plt.title("Image from URL")
plt.show()
```

Perform Multiple Transformations

a. Resize an image to 256x256 pixels

```python
import cv2

# Assuming 'img_rgb' is your image
resized_image = cv2.resize(img_rgb, (256, 256))

# Display the resized image (optional)
plt.imshow(resized_image)
plt.axis('off')
plt.title("Resized Image (256x256)")
plt.show()
```

b. Rotate by 180 degrees clockwise.

```python
import cv2

# Assuming 'img_rgb' is your image
rotated_image = cv2.rotate(img_rgb, cv2.ROTATE_180)

# Display the rotated image (optional)
plt.imshow(rotated_image)
plt.axis('off')
plt.title("Rotated Image (180 degrees)")
plt.show()
```

c. Apply a vertical flip.



```python
import cv2

# Assuming 'img_rgb' is your image
flipped_image = cv2.flip(img_rgb, 0)  # 0 represents vertical flip

# Display the flipped image (optional)
plt.imshow(flipped_image)
plt.axis('off')
plt.title("Vertically Flipped Image")
plt.show()
```

d. Display all transformations in a single subplot.

```python
import matplotlib.pyplot as plt
import cv2

# Assuming 'img_rgb' is your original image

# Resize
resized_image = cv2.resize(img_rgb, (256, 256))

# Rotate
rotated_image = cv2.rotate(img_rgb, cv2.ROTATE_180)

# Flip
flipped_image = cv2.flip(img_rgb, 0)  # 0 for vertical flip

# Create a figure and subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Display the original image
axes[0, 0].imshow(img_rgb)
axes[0, 0].set_title("Original Image")
axes[0, 0].axis('off')

# Display the resized image
axes[0, 1].imshow(resized_image)
axes[0, 1].set_title("Resized Image")
axes[0, 1].axis('off')

# Display the rotated image
axes[1, 0].imshow(rotated_image)
axes[1, 0].set_title("Rotated Image")
axes[1, 0].axis('off')

# Display the flipped image
axes[1, 1].imshow(flipped_image)
axes[1, 1].set_title("Flipped Image")
axes[1, 1].axis('off')

# Adjust layout to prevent overlapping titles
plt.tight_layout()

# Show the subplots
plt.show()
```
