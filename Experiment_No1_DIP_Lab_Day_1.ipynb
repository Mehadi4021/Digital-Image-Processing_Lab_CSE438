## 1. a)	Write a program to read, display and write an image. [From Drive]

```python
# Step 1: Provide the image path in your Google Drive
# Replace with the actual path to your image file
image_path = '/content/drive/MyDrive/Digital Image Processing_Lab/Lab_1/Picture.png'
```

```python
# Step 2: Read the image using OpenCV
img = cv2.imread(image_path)  # OpenCV reads images in BGR format
```

```python
# Step 3: Convert BGR to RGB for correct display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# Step 4: Display the image using Matplotlib
plt.imshow(img_rgb)
plt.axis('off')  # Remove axes for cleaner display
plt.title("Image from Google Drive")
plt.show()
```

```python
# Step 5: Write/Save the image to a new file
output_path = '/content/drive/MyDrive/Digital Image Processing_Lab/Lab_1/output.png'
cv2.imwrite(output_path, img)
print(f"Image saved at: {output_path}")
```

## 1. b)	Write a program to read, write and display an image. [From URL]

```python
# Step 1: Provide the URL of the image
image_url = 'https://images.unsplash.com/photo-1462536943532-57a629f6cc60?fm=jpg&q=60&w=3000&ixlib=rb-4.0.3'  # Replace with the actual URL
```

```python
# Step 2: Download the image using urllib
with urllib.request.urlopen(image_url) as response:
    img_array = np.asarray(bytearray(response.read()), dtype=np.uint8)  # Convert to NumPy array
```

```python
# Step 3: Decode the image (convert it into OpenCV format)
img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)  # Decode as a color image (BGR format)
```

```python
# Step 4: Convert BGR to RGB for correct display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python

# Step 5: Display the image using Matplotlib
plt.imshow(img_rgb)
plt.axis('off')  # Remove axes for cleaner display
plt.title("Image from URL")
plt.show()
```

```python
# Step 6: Save the image locally
output_path = '/content/drive/MyDrive/Digital Image Processing_Lab/Lab_1/friendship.jpg'
cv2.imwrite(output_path, img)
print(f"Image saved as: {output_path}")
```
