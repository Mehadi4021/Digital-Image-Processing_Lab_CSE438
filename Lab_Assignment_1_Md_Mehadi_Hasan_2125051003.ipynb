```python
from google.colab import files
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import urllib.request
import numpy as np
```

```python
# Step 1: Provide the URL of the image
image_url = 'https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcREoRGyXmHy_6aIgXYqWHdOT3KjfmnuSyxypw&s'
```

```python
# Step 2: Download the image using urllib
with urllib.request.urlopen(image_url) as response:
    img_array = np.asarray(bytearray(response.read()), dtype=np.uint8)  # Convert to NumPy array
```

```python
# Step 3: Decode the image (convert it into OpenCV format)
img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)  # Decode as a color image (BGR format)
```

```python
# Step 4: Convert BGR to RGB for correct display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# Step 5: Display the image using Matplotlib
plt.imshow(img_rgb)
plt.axis('off')  # Remove axes for cleaner display
plt.title("Image from URL")
plt.show()
```

Task-01: Extract a 100100 Region from the Center of the Image read from an URL.

```python
img_crop=img_rgb[0:100,0:100]

plt.imshow(img_crop)
plt.axis('off')
plt.title("Cropped 100100 Region")
plt.show()
```

Task 2:  Convert the image to HSV color space, visualize each components in grayscale representation.

```python
hsv_image = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h_channel, s_channel, v_channel = cv2.split(hsv_image)
```

```python
plt.imshow(h_channel, cmap="gray")
plt.axis("off")
plt.title("Hue Channel (H)")
plt.show()


plt.imshow(v_channel, cmap="gray")
plt.title("Value Channel (V)")
plt.axis("off")
plt.show()


plt.imshow(s_channel, cmap="gray")
plt.title("Saturation Channel (S)")
plt.axis("off")
plt.show()
```
