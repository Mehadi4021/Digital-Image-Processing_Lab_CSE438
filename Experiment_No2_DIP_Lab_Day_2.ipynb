Import necessary libraries

```python
from google.colab import files
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import urllib.request
import numpy as np
```

#### 1. a) Read the image from the URL

```python
# Step 1: Provide the URL of the image
image_url = 'https://cdn.shopify.com/s/files/1/0274/9503/9079/files/20220211142754-margherita-9920_5a73220e-4a1a-4d33-b38f-26e98e3cd986.jpg?v=1723650067'  # Replace with the actual URL
```

```python
# Step 2: Download the image using urllib
with urllib.request.urlopen(image_url) as response:
    img_array = np.asarray(bytearray(response.read()), dtype=np.uint8)  # Convert to NumPy array
```

```python
# Step 3: Decode the image (convert it into OpenCV format)
img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)  # Decode as a color image (BGR format)
```

```python
# Step 4: Convert BGR to RGB for correct display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python

# Step 5: Display the image using Matplotlib
plt.imshow(img_rgb)
plt.axis('off')  # Remove axes for cleaner display
plt.title("Image from URL")
plt.show()
```

## 1. b)	Display the properties of the given image like height, width, no. of channels and size.

```python
# Step 1: Get the image properties
height, width, channels = img.shape  # Height, width, and number of color channels
size = img.size  # Total number of pixels (width × height × channels)
```

```python
# Step 2: Print the properties
print("Image Properties:")
print(f"Height: {height} pixels")
print(f"Width: {width} pixels")
print(f"Number of Channels: {channels}")
print(f"Size: {size} bytes")  # Total size in bytes
```

## 1. c)	Display the individual channels (R, G, B) pixel values in that image.

```python
# Step 1: Split the image into its B, G, R channels
blue_channel, green_channel, red_channel = cv2.split(img)
```

```python
# Step 2: Display the pixel values of each channel
print("Blue Channel Pixel Values:")
print(blue_channel)

print("\nGreen Channel Pixel Values:")
print(green_channel)

print("\nRed Channel Pixel Values:")
print(red_channel)
```

## 1.d) Separate the given color image in three R G & B color channels.

```python
# Step 1: Visualize the individual channels
# Convert each channel to grayscale for display
plt.figure(figsize=(10, 8))

plt.subplot(1, 3, 1)
plt.title("Blue Channel")
plt.imshow(blue_channel,cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("Green Channel")
plt.imshow(green_channel,cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("Red Channel")
plt.imshow(red_channel,cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## 1.e) Convert the given color image into gray-scale image and display the shape of the gray-scale image.

```python
# Step 1: Convert the image to grayscale
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Step 2: Display the shape of the grayscale image
print(f"Shape of the Grayscale Image: {gray_image.shape}")

# Step 3: Visualize the grayscale image
plt.figure(figsize=(6, 6))
plt.title("Grayscale Image")
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.show()
```
