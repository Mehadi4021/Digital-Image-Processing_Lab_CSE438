

## üß† Concept: What is Spatial Resolution?

**Spatial resolution** refers to the **number of pixels** used to represent an image.  
- Higher resolution = More pixels = More detail  
- Lower resolution = Fewer pixels = Less detail, more blockiness or blurriness

---

## üìâ What Happens When You Reduce Resolution?

1. **Loss of detail** ‚Äì Fine structures vanish.
2. **Increased pixelation** ‚Äì Image looks blocky.


---

## üéì Classroom-Friendly Python Code (with Illustration)

Here‚Äôs a full Jupyter Notebook-style code that demonstrates the effects:

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
img = cv2.imread('your_image.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB

# Define different resolutions
scales = [1.0, 0.5, 0.25, 0.125]  # 100%, 50%, 25%, 12.5%

# Prepare plot
plt.figure(figsize=(12, 8))
for i, scale in enumerate(scales):
    # Resize down
    small = cv2.resize(img, (0, 0), fx=scale, fy=scale, interpolation=cv2.INTER_AREA)
    # Resize back up to original for display
    upscaled = cv2.resize(small, (img.shape[1], img.shape[0]), interpolation=cv2.INTER_NEAREST)
    
    # Plot
    plt.subplot(2, 2, i+1)
    plt.imshow(upscaled)
    plt.title(f'Resolution: {int(scale*100)}%')
    plt.axis('off')

plt.tight_layout()
plt.suptitle("üîç Effects of Reducing Spatial Resolution", fontsize=16, y=1.02)
plt.show()
```

---

| Resolution | Effects |

| 50%          | Slight blur, minor loss of detail |


## üéØ What Are Intensity Levels?

In grayscale images:
- Intensity levels refer to the **range of brightness values**.
- A standard 8-bit image has **256 levels (0 to 255)**, where:
  - 0 = Black
  - 255 = White

---

## üìâ Effects of Reducing Intensity Levels

| Levels | Bits | Visual Effect |
|--------|------|----------------|
| 256    | 8    | Smooth transitions, full detail |
| 128    | 7    | Slight banding begins |
| 64     | 6    | Visible posterization |
| 16     | 4    | Loss of fine detail, harsh transitions |
| 8      | 3    | Very blocky look |
| 2      | 1    | Only black and white (binary image) |

---

## üß™ Python Demo (Classroom Notebook Style)

Here‚Äôs a code snippet for a class-friendly demonstration:

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load grayscale image
img = cv2.imread('your_image.jpg', cv2.IMREAD_GRAYSCALE)

# Intensity levels to test
levels = [256, 128, 64, 16, 8, 2]

# Plot results
plt.figure(figsize=(15, 8))
for i, level in enumerate(levels):
    # Reduce intensity levels
    quantized = np.floor(img / (256 / level)) * (256 / level)
    quantized = quantized.astype(np.uint8)
    
    # Show image
    plt.subplot(2, 3, i+1)
    plt.imshow(quantized, cmap='gray')
    plt.title(f'{level} Levels')
    plt.axis('off')

plt.tight_layout()
plt.suptitle("üé® Effects of Reducing Intensity Levels", fontsize=16, y=1.05)
plt.show()
```

---

## üß† What Students Should Notice:

- **Banding** or **posterization** as levels go down.
- **Smooth grayscale transitions disappear**.
- Fewer levels = **less storage**, but **less visual quality**.


````python
# Load grayscale image
img = cv2.imread('/content/drive/MyDrive/DIP_Lab/Day_7/man_with_camera.jpg', cv2.IMREAD_GRAYSCALE)
````

````python
# Intensity levels to test
levels = [256, 128, 64, 16, 8, 2]
````

````python
# Plot results
plt.figure(figsize=(15, 8))
for i, level in enumerate(levels):
    # Reduce intensity levels
    quantized = np.floor(img / (256 / level)) * (256 / level)
    quantized = quantized.astype(np.uint8)

    # Show image
    plt.subplot(2, 3, i+1)
    plt.imshow(quantized, cmap='gray')
    plt.title(f'{level} Levels')
    plt.axis('off')

plt.tight_layout()
plt.suptitle("üé® Effects of Reducing Intensity Levels", fontsize=16, y = 1.05)
plt.show()
````
